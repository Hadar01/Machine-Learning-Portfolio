# **<h1>2.Bivariate Analysis<h1>**
In bivariate analysis we perform the analysis on multiple columns to get more information on data


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
titanic=pd.read_csv('train.csv')
petals=sns.load_dataset('iris')
flights=sns.load_dataset('flights')

```

# **ScatterPlot for multiple column analysis(NUMERICAL VS NUMERICAL)**
By using hue,style and size , Ill display multiple information from the dataset


```python
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue=tips['smoker'],style=tips['sex'],size=tips['size'])
```

```python
This shows a scatter plot of total bill vs tip, with the size of the points representing the size of the party,
 the color representing whether the party was smoking or not, 
 and the style representing 
```

# **BARPlot for multiple column analysis(NUMERICAL VS CATEGORICAL)**

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
This shows us that the average total bill is highest on Sunday and lowest on Friday.
```

# **BoxPlot for multiple column analysis(NUMERICAL VS CATEGORICAL)**

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
This shows us that the average total bill is higher for Males than Females.
```

# **DistPlot for multiple column analysis(NUMERICAL VS CATEGORICAL)**

```python
sns.histplot(x='total_bill',data=tips,hue='smoker',color='red')
```

# 
**BoxPlot for multiple column analysis(categorical VS CATEGORICAL)**

```python
pd.crosstab(tips['smoker'], tips['sex'])
```

```python
sns.clustermap(pd.crosstab(tips['smoker'], tips['sex']),cmap='coolwarm',annot=True,linewidths=1)
```

```python
sns.heatmap(pd.crosstab(tips['smoker'], tips['sex']),cmap='coolwarm',annot=True,linewidths=1)
```

# 
**PairPlot for multiple column analysis**
It automatically detects all the numerical columns between the datasets and plots multiple scatterplot among them

```python
sns.pairplot(petals,hue='species',palette='coolwarm')
```

# 
**LinePlot for multiple column analysis(NUMERICAL)**
It is used when we have to plot (time vs numerical)

```python
flights.head()
```

```python
sns.lineplot(x='year', y='passengers', data=flights)
```

```python
sns.heatmap(flights.pivot_table(index='month',columns='year',values='passengers'),cmap='coolwarm',linecolor='white',linewidth=1)
```

Exploaratory Data Analysis
It basically of 2 types :
1.Univariate Analysis
2.Bivariate Analysis




Univariate Analysis :-
                        Where we work on uni or single column of data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

1. Categorical Data
The data which can be categorised in certain types such as true ,false etc

```python

sns.countplot(x='Survived',data=df)

```

```python
sns.countplot(x='Pclass',data=df,hue='Survived')
```

```python
sns.countplot(x='Sex',data=df)
```

```python
sns.countplot(x='Embarked',data=df)
```

```python
plt.pie(df['Survived'].value_counts(),labels=['Not Survived','Survived'],autopct='%1.1f%%')
```

We can also use pie chart if we want to compare the data through percentage

2.Numerical data


```python
plt.hist(df['Age'],bins=30)
```

```python
sns.displot(df['Age'],kde=False,bins=30)
```

```python
sns.displot(df['Age'],kde=True,bins=30)
```

The line is known as kde (kernel density estimation)

```python
sns.boxplot(x='Fare',data=df)
```

Boxplot is used for getting the outliers and also finding maximum and minimum in the data
